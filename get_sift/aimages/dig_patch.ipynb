```python
import cv2
import numpy as np

# 图片路径
image_path = '/media/human_face_need_test/PsPDv3/064/2/left_up06401010000.bmp'

# 读取图像
img = cv2.imread(image_path)

# 确保图像是灰度图，如果是彩色图则转换为灰度图
if img.ndim == 3:
    img = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)

# 图像的行数和列数
num_rows = 60  # 总行数
patch_size = 64  # 每个patch的大小

# 要提取的patch的行和列索引
i = 15  # 第i行
k = 3  # 第k列

# 计算patch的起始坐标
start_row = (i - 1) * patch_size
start_col = (k - 1) * patch_size

# 计算patch的结束坐标
end_row = start_row + patch_size
end_col = start_col + patch_size

# 提取patch
patch = img[start_row:end_row, start_col:end_col]

# 显示提取的patch

# 保存提取的patch
patch_filename = f'patch_row_{i}_col_{k}.png'
cv2.imwrite(patch_filename, patch)

print(f'Patch saved as {patch_filename}')
```
