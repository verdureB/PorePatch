```python
import cv2
import os
def extract_frame(video_path, frame_number, output_path):
    # 打开视频文件
    cap = cv2.VideoCapture(video_path)
    
    # 获取视频的总帧数
    total_frames = int(cap.get(cv2.CAP_PROP_FRAME_COUNT))
    
    # 检查请求的帧数是否在有效范围内
    if frame_number >= total_frames:
        print(f"Frame number {frame_number} is out of range. Total frames: {total_frames}")
        return
    
    # 设置视频的当前帧位置
    cap.set(cv2.CAP_PROP_POS_FRAMES, frame_number)
    
    # 读取当前帧
    ret, frame = cap.read()
    
    # 检查是否成功读取帧
    if ret:
        # 保存帧为图像文件
        cv2.imwrite(output_path, frame)
        print(f"Frame {frame_number} saved as {output_path}")
    else:
        print(f"Failed to read frame {frame_number}")
    
    # 释放视频捕获对象
    cap.release()
```

```python

# 我们所需提取帧的视频所在的文件夹位置
mp4_data_root ='/media/dataset/origin_eth_eys'
# 所有人
all_human = sorted([i for i in os.listdir(mp4_data_root)  if  os.path.isdir(os.path.join(mp4_data_root, i))])

#实际上使用的表情
true_emotion = 'EXP-2-eyes'
# 为了方便后续文件，不需要改的命名
name_emotion = 'EMO-1-shout+laugh'
# 提取出来的帧保存的文件夹位置
frame_data_store_root= '/media/human_face_need_test/PsPD'

# 提取出视频的哪一帧,为了统一命名，方便后续文件不需要改的
name_virtual_pick_frames_number = '2'
# 实际上选取的帧数
true_pick ='1'
```
