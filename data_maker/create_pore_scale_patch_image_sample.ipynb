```python
from PIL import Image

def stitch_patches(image_path, output_path, rows_to_extract, patch_size=64, num_rows=60):
    """
    从拼接图像中提取指定行数的patch，并将它们拼接成一张新图像。

    Args:
        image_path: 输入图像路径。
        output_path: 输出图像路径。
        rows_to_extract: 要提取的行数列表（从倒数第几行开始计数，例如[7, 3]表示倒数第7行和倒数第3行）。
        patch_size: patch的大小 (默认为64x64)。
        num_rows: 拼接图像的行数 (默认为60)。
    """
    try:
        img = Image.open(image_path)
        width, height = img.size

        # 计算每行patch的数量
        num_cols = width // patch_size
        num_cols =8
        # 计算要提取的行的索引（从0开始）
        row_indices = [num_rows - row for row in rows_to_extract]

        # 创建新图像, 高度根据要提取的行数确定
        new_img_height = len(rows_to_extract) * patch_size
        new_img = Image.new('RGB', (num_cols * patch_size, new_img_height))

        # 提取patch并拼接
        for i, row_index in enumerate(row_indices):
            for j in range(num_cols):
            #for j in range(8):
                left = j * patch_size
                top = row_index * patch_size
                right = left + patch_size
                bottom = top + patch_size
                patch = img.crop((left, top, right, bottom))

                # 计算patch在新图像中的位置
                paste_x = left
                paste_y = i * patch_size # 根据行索引计算垂直位置
                new_img.paste(patch, (paste_x, paste_y))

        new_img.save(output_path)
        print(f"拼接后的图像已保存到: {output_path}")

    except FileNotFoundError:
        print(f"错误: 找不到图像文件: {image_path}")
    except Exception as e:
        print(f"发生错误: {e}")


# 示例用法:
image_path = "input.jpg"  # 替换为你的图像路径
output_path = "create_pore_scale1.jpg" # 替换为你的输出路径
rows_to_extract = [26, 24, 25] #  提取倒数第7行、倒数第3行和倒数第2行
stitch_patches(eye_eyebrow, output_path, rows_to_extract)
```

```python
from PIL import Image

def vertically_stitch_images(image_path1, image_path2, output_path):
    """
    将两张图像垂直拼接到一起。

    Args:
        image_path1: 第一张图像的路径。
        image_path2: 第二张图像的路径。
        output_path: 输出图像的路径。
    """
    try:
        img1 = Image.open(image_path1)
        img2 = Image.open(image_path2)

        # 获取图像的宽度和高度
        width1, height1 = img1.size
        width2, height2 = img2.size

        # 确定新图像的宽度（取最大宽度）
        new_width = max(width1, width2)

        # 创建一个新的空白图像，用于容纳拼接后的图像
        new_img = Image.new('RGB', (new_width, height1 + height2))

        # 将第一张图像粘贴到新图像的顶部
        new_img.paste(img1, (0, 0))

        # 将第二张图像粘贴到第一张图像的下方
        new_img.paste(img2, (0, height1))

        # 保存拼接后的图像
        new_img.save(output_path)
        print(f"拼接后的图像已保存到: {output_path}")


    except FileNotFoundError:
        print(f"错误：找不到图像文件：{image_path1} 或 {image_path2}")
    except Exception as e:
        print(f"发生错误：{e}")


# 示例用法
image_path1 = "/media/human_face_need_test/brown_v2/create_pore_scale1.jpg"  # 替换为你的第一张图像路径
image_path2 = "/media/human_face_need_test/brown_v2/create_pore_scale2.jpg"  # 替换为你的第二张图像路径
output_path = "/media/human_face_need_test/brown_v2/create_pore_scale_aLL.jpg"  # 替换为你的输出路径

vertically_stitch_images(image_path1, image_path2, output_path)
```
